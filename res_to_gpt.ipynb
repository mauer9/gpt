```python
from pymilvus import MilvusClient
from langchain.embeddings import OpenAIEmbeddings
```

```python
import os
os.environ['OPENAI_API_KEY'] = '***'
```

```python
client = MilvusClient(
    uri='***', # Cluster endpoint obtained from the console
    token='***'
)
```

```python
query = 'Коротко опиши основные правила для банков второго уровня'
```

```python
embedding = OpenAIEmbeddings()
vector = embedding.embed_query(query)
```

```python
a1_res =  client.search('a1', [vector], output_fields=['text'])[0]
```

```python
a2_res = client.search('a2', [vector], output_fields=['text'])[0]
```

```python
a1_res1 = [v['entity']['text'] for v in a1_res]
```

```python
a2_res1 = [v['entity']['text'] for v in a2_res]
```
